```python
# Check data type
df_ords_prods_merge['user_id'].dtype
```

```python
# Changing variable's (user_id) data type 
df_ords_prods_merge['user_id'] = df_ords_prods_merge['user_id'].astype('str')
```

```python
# Re_Check data type
df_ords_prods_merge['user_id'].dtype
```

```python
# Step 4 - Create column 'max_order', applaying the transform() function on the 'order_number' column and spliting data into groups based on the 'user_id' column
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# Step 4 - Deriving columns with loc ()
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Frequency of 'loyalty_flag' column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# Step 5 - Check the basic statistics of the product prices for each loyalty category 
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

```python
# Check for prices above $100
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 100]
```

```python
# Replace outliers with missing values
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# Check max price
df_ords_prods_merge['prices'].max()
```

```python
# Step 5 - Re-check the basic statistics of the product prices for each loyalty category 
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

```python
# Step 6 - Create column 'average_price', applaying the transform() function on the 'prices' column and spliting data into groups based on the 'user_id' column
df_ords_prods_merge['average_price']=df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Step 6 - Deriving columns with loc ()
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python

df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Frequency of 'spending_flag' column
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.head(10)
```

```python
# Step 7 - Create column 'median__days_since_prior_order', applaying the transform() function on the 'days_since_prior_order' column and spliting data into groups based on the 'user_id' column
df_ords_prods_merge['median__days_since_prior_order']=df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head(10)
```

```python
# Step 7 - Deriving columns with loc ()
df_ords_prods_merge.loc[df_ords_prods_merge['median__days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median__days_since_prior_order'] <= 20) & (df_ords_prods_merge['median__days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python

df_ords_prods_merge.loc[df_ords_prods_merge['median__days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Frequency of 'order_frequency_flag' column
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check the accuracy of the 'order_frequency_flag'
df_ords_prods_merge[['user_id', 'order_frequency_flag', 'days_since_prior_order', 'median__days_since_prior_order', 'order_number']].head(15)
```
