# 01. Importing libraries

```python
# Step 1 - Import dataframes_merged dataframe using the os library
df_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'dataframes_merged.pkl'))
```

# 03. Creating charts

```python
# Step 2 - Create bar chart
bar1=df_merged['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Sort bars according to their index order
bar2=df_merged['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Step 2 - Create histogram
hist=df_merged['prices'].plot.hist(bins=25)
```

```python
# Increase the number of bins
hist2= df_merged['prices'].plot.hist(bins=70)
```

```python
# Step 2 - Create scatterplot
scatterplot=sns.scatterplot(x = 'prices', y = 'prices', data = df_merged)
```

```python
# Create sample
np.random.seed(4) 
dev = np.random.rand(len(df_merged))<= 0.7
```

```python
dev
```

```python
np.random.rand(10)
```

```python
# Store 70% of the sample in the df big
big=df_merged[dev]
```

```python
# Store 30% of the sample in the df small
small=df_merged[~dev]
```

```python
# Check numbers of rows of entire df
len(df_merged)
```

```python
# Check numbers of rows 
len(big)+len(small)
```

```python
# Reduce sample to only those columns necessary for chart by creating new df
df_2=small[['orders_day_of_week', 'prices']]
```

```python
# Step 2 - Create line chart
line=sns.lineplot(data=df_2, x='orders_day_of_week', y='prices')
```

```python
# Step 3 - Create histogram of the 'order hour of day' column
hist_order_hour_of_day=df_merged['order_hour_of_day'].plot.hist(bins=24)
```

The histogram shows every order hour of the day in the entire 'df merge' data frame. So we can see that the most orders were in the hours between 10-16.

```python
# Step 4 - Create bar chart from the 'loyalty flag' column
bar_loyalty_flag=df_merged['loyalty_flag'].value_counts().plot.bar()
```

```python
# Step 5 - Reduce sample to only those columns necessary for chart by creating new df
df_line=small[['order_hour_of_day', 'prices']]
```

```python
# Step 5 - Create line chart
line_hour_price=sns.lineplot(data=df_line, x='order_hour_of_day', y='prices')
```

```python
# Step 6 -Reduce sample to only those columns necessary for chart by creating new df
df_line2=small[['age', 'number_of_dependants']]
```

```python
# Step 6 - Create line chart
line_age_num_of_dep=sns.lineplot(data=df_line2, x='age', y='number_of_dependants')
```

Line charts shows variable 'age' on x-axis and variable 'number of dependents' on y-axis.

```python
# Step 7 - Create scatterplot
scatterplot_age_spending_power=sns.scatterplot(x = 'age', y = 'income', data = df_merged)
```

The scatterplot shows the variable 'age' on the x-axis and the variable 'income' on the y-axis. It shows that the presented age groups mostly earn below 200,000, and that the earnings are higher in age groups over 40 years old.
